# Distribution of Crashes based on 'Weather'

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a file path
road_crash_data = Path("nsw_road_crash_data.csv")
# Read in the data. 
road_crash_data_df = pd.read_csv(road_crash_data)
road_crash_data_df.head()
```

```python
road_crash_data_df.columns.unique()
```

```python
road_crash_data_df["Weather"].unique()
```

# Bar graph displaying Distribution of Crashes by Weather Type

```python
# Bar graph displaying Distribution of Crashes by Weather Type
weather_counts = road_crash_data_df['Weather'].value_counts()

plt.figure(figsize=(10, 6))
weather_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Weather Type')
plt.ylabel('Number of Crashes')
plt.title('Number of Crashes in Each Weather Type')
plt.xticks(rotation=45)
plt.show()
```

# Pie chart displaying Distribution of Crashes by Weather Type

```python
# Pie chart displaying Distribution of Crashes by Weather Type
plt.figure(figsize=(8, 8))
plt.pie(weather_counts, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.legend(weather_counts.index, title="Weather Type", bbox_to_anchor=(1, 0.5), loc="center left")
plt.title('Distribution of Crashes by Weather Type')
plt.show()
```
