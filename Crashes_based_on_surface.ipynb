# Distribution of crashed based on 'Surface Condition' and 'Road Surface

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a file path
road_crash_data = Path("nsw_road_crash_data.csv")
# Read in the data. 
road_crash_data_df = pd.read_csv(road_crash_data)
road_crash_data_df.head()
```

```python
road_crash_data_df["Road surface"].unique()
```

```python
road_crash_data_df["Surface condition"].unique()
```

# Pie chart displaying Distribution of Crashes by Surface Condition

```python
condition_counts = road_crash_data_df['Surface condition'].value_counts()

# Pie chart displaying Distribution of Crashes by Weather Type
plt.figure(figsize=(8, 8))
plt.pie(condition_counts, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.legend(condition_counts.index, title="Surface Condition", bbox_to_anchor=(1, 0.5), loc="center left")
plt.title('Distribution of Crashes by Surface Condition')
plt.show()
```

# Pie chart displaying Distribution of Crashes by Road surface

```python
# Bar graph displaying Distribution of Crashes by Road Surface
surface_counts = road_crash_data_df['Road surface'].value_counts()

# Pie chart displaying Distribution of Crashes by Weather Type
plt.figure(figsize=(8, 8))
plt.pie(surface_counts, autopct='%1.1f%%', startangle=140)
plt.legend(surface_counts.index, title="Road Surface", bbox_to_anchor=(1, 0.5), loc="center left")
plt.title('Distribution of Crashes by Road Surface')
plt.show()
```
